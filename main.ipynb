```python
import pandas as pd
import os
```

# 1. Importando dados de empréstimos a partir dos datasets em .csv

```python
pasta = 'Datasets/dados_emprestimos'
arquivos = [f for f in os.listdir(pasta) if ".csv" in f and os.path.isfile(os.path.join(pasta, f))]
df=pd.DataFrame()

for arquivo in arquivos:
    print(arquivo)
    arquivo="Datasets/dados_emprestimos/"+arquivo
    df = pd.concat([df, pd.read_csv(arquivo)], axis=0, ignore_index=True)
```
